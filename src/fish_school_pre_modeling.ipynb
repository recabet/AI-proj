```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../data/school_table_thresh.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns = ['number','file_name', 'label', 'bbox', 'gps_lon_lat', 'is_very_wide','is_very_tall','dis_to_surface', 'dis_level', 'intensity_img','width_length_ratio', 'axis_ellipse_ratio','perimeter_area_ratio','dis_to_bottom'])
df.head()
```

```python
def centeroid(df):
    df[['center_x', 'center_y']] = df['center'].str.strip('()').str.split(',', expand=True)
    df[['center_x', 'center_y']] = df[['center_x', 'center_y']].astype(float)
    return df
df = centeroid(df)
df
```

```python
df = df.drop(columns = ['center'])
df.head()
```

```python
def time(df):
    df[['date','time1']] = df['time'].str.split(' ',expand=True)
    df[['year','month','day']] = df['date'].str.split('-',expand=True)
    df['month'] = df['month'].astype(int)
    df[['hour','minute','second']] = df['time1'].str.split(':',expand = True)
    df['hour'] = df['hour'].astype(int)
    return df
df = time(df)
df = df.drop(columns = ['year','day','second','minute','date','time1','month','time'])
df.head()
```

```python
def night(df):
    df['night'] = ((df['hour'] < 5) | (df['hour'] > 18)).astype(int)
    return df

df = night(df)
df.head()
```

```python
df = df.drop(columns = 'hour')
df
```

```python
df['area_school'] = (df['axis_major_length']) * (df['axis_minor_length']) * np.pi
df.head()
```

```python
df = df.drop(columns = ['axis_major_length','axis_minor_length'])
df.head()
```

```python
df.to_csv('fish_school_pre_modeling.csv')
```
